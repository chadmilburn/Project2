```python
# import depend
import pandas as pd
```

```python
# Create a reference to the CSV and import it into a Pandas DataFrame
csv_path = 'Datafiniti_Hotel_Reviews.csv'
df = pd.read_csv(csv_path)
df.head()
```

```python
# drop all rows not in the city of Las Vegas
index_names = df[df['city']!= 'Las Vegas'].index
df.drop(index_names, inplace=True)
df.head()
```

```python
#get list of column names
column_names = df.columns.values.tolist()
column_names
```

```python
# drop unneeded columns
df = df.drop(['id',
 'dateAdded',
 'dateUpdated',
 'primaryCategories',
 'country',
 'keys',
 'reviews.dateSeen',
 'reviews.sourceURLs',
 'reviews.title',
 'reviews.userCity',
 'reviews.userProvince',
 'reviews.username',
 'sourceURLs',
 'websites'], axis=1)
df.reset_index(inplace=True, drop=True)
df.head()
```

```python
#save cleaned csv for later 
df.to_csv("Datafiniti_Hotel_Review_cleaned.csv", index=False)
```

```python
import pandas as pd
```

```python
csv_path = 'LasVegasTripAdvisorReviews-Dataset.csv'
df = pd.read_csv(csv_path, delimiter=';')
df.head()
```

```python
#get list of column names
column_names = df.columns.values.tolist()
column_names
```

```python
# drop unneeded columns
df = df.drop([ 
    'Nr. reviews',
    'Nr. hotel reviews',
    'Helpful votes',
    'Period of stay',
    'Gym',
    'Tennis court',
    'Spa',
    'Free internet',
    'Hotel stars',
    'Nr. rooms',
    'User continent',
    'Member years',
    'Review weekday'], axis=1)
df.reset_index(inplace=True, drop=True)
df.head()
```

```python
#save cleaned csv for later 
df.to_csv("LV_TA_cleaned.csv", index=False)
```

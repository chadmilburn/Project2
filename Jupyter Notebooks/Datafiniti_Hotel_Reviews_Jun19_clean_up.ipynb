```python
# import depend
import pandas as pd
import datetime as dt
import calendar
```

```python
# Create a reference to the CSV and import it into a Pandas DataFrame
csv_path = 'Datafiniti_Hotel_Reviews_Jun19.csv'
df = pd.read_csv(csv_path)
df.head()
```

```python
# drop all rows not in the city of Las Vegas
index_names = df[df['city']!= 'Las Vegas'].index
df.drop(index_names, inplace=True)
df.head()
```

```python
#get list of column names
column_names = df.columns.values.tolist()
#column_names
```

```python
# drop unneeded columns
df = df.drop(['id',
 'dateAdded',
 'dateUpdated',
 'primaryCategories',
 'country',
 'keys',
 'reviews.dateSeen',
 'reviews.dateAdded',
 'reviews.sourceURLs',
 'reviews.title',
 'reviews.userCity',
 'reviews.username',
 'sourceURLs',
 'websites'], axis=1)
df.reset_index(inplace=True, drop=True)
df.head()
```

```python
# clean date column
df.astype({'reviews.date' :'datetime64[ns]' })
```

```python
# convert review date column to date time
df['reviews.date'] = pd.to_datetime(df['reviews.date'])
```

```python
# extract month
df['reviews.date'] = df['reviews.date'].dt.month
```

```python
# set month int to full name
df['reviews.date'] = df['reviews.date'].apply(lambda x: calendar.month_name[x])
```

```python
df.head()
```

```python
#save cleaned csv for later 
df.to_csv("Datafiniti_Hotel_Reviews_Jun19_cleaned.csv", index=False)
```

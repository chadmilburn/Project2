```python
import pandas as pd
```

```python
csv_path = '7282_1_cleaned.csv'
reviewsfile1 = pd.read_csv(csv_path)
reviewsfile1.head()
```

```python
len(reviewsfile1)
```

```python
csv_path2 = 'Datafiniti_Hotel_Review_cleaned.csv'
reviewsfile2=pd.read_csv(csv_path2)
reviewsfile2.head()
```

```python
len(reviewsfile2)
```

```python
csv_path2 = 'Datafiniti_Hotel_Reviews_Jun19_cleaned.csv'
reviewsfile3=pd.read_csv(csv_path2)
reviewsfile3.head()
```

```python
len(reviewsfile3)
```

```python
#concat files into 1
combined_file_1 = pd.concat([reviewsfile1, reviewsfile2, reviewsfile3], axis =0)
combined_file_1.head(5)
```

```python
len(combined_file_1)
```

```python
combined_file_1.head()
```

```python
column_names = combined_file_1.columns.values.tolist()
column_names
```

```python
combined_file_1.rename(columns = {'reviews.date':'reviews_date',
                                 'reviews.rating': 'reviews_rating',
                                 'reviews.text': 'reviews_text',
                                 'reviews.userProvince': 'reviews_userProvince'}, inplace = True) 
```
